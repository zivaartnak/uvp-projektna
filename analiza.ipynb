```python
import pandas as pd
clanki = pd.read_csv('obdelani_podatki/clanki.csv')
clanki
```

```python
import matplotlib.pyplot as plt

clanki['avtor'].value_counts().plot.pie()
plt.axis('equal')
plt.show()
```

```python
clanki['word_count'] = clanki['vsebina'].apply(lambda x: len(x.split()))
average_word_count = clanki.groupby('avtor')['word_count'].mean()

average_word_count.plot(kind='bar')
plt.xlabel('Avtor')
plt.ylabel('Povprečno število besed')
plt.title('Povprečno število besed v članku po avtorjih')
plt.show()
```

```python
clanki['title_length'] = clanki['naslov'].apply(lambda x: len(x.split()))
average_title_length = clanki.groupby('avtor')['title_length'].mean()

average_title_length.plot(kind='bar')
plt.xlabel('Avtor')
plt.ylabel('Povprečna dolžina naslova')
plt.title('Povprečna dolžina naslova po avtorjih')
plt.show()
```
